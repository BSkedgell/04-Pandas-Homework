```python
##TOP SPENDERS
# Calculate purchase count for each SN
count_purchase = purchase_data.groupby('SN')['Price'].count()
count_per_SN_df = pd.DataFrame(count_purchase)

# Calculate total purchase price for each SN
total_pp_SN_df = purchase_data.groupby('SN')['Price'].sum()

# Calculate average purchase price
avg_pp_SN = purchase_data.groupby('SN')['Price'].mean()

# Add Average Purchase Price to table
count_per_SN_df['Average Purchase Price'] = avg_pp_SN

# Add Total Purchase Value to table
count_per_SN_df['Total Purchase Value'] = total_pp_SN_df
count_per_SN_df

# Rename price to purchase count
updated_count_per_SN_df = count_per_SN_df.rename(columns={
    'Price': 'Purchase Count',
})

# Sort price desc and format $ to last two columns
sorted_SN_df = updated_count_per_SN_df.sort_values(by=['Total Purchase Value'], ascending=False)
sorted_SN_df['Average Purchase Price'] = count_per_SN_df['Average Purchase Price'].map("${:.2f}".format)
sorted_SN_df['Total Purchase Value'] = count_per_SN_df['Total Purchase Value'].map("${:.2f}".format)
sorted_SN_df

# Show top 5 spenders only
sorted_SN_df.head(5)

```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
##MOST POPULAR ITEMS
#Calculate & variable names
grp_item = purchase_data.groupby(["Item ID", "Item Name"])
grp_price_sum = grp_item["Price"].sum()
grp_item_1 = grp_item["SN"].count()
grp_item_2 = (grp_price_sum / grp_item_1)
grp_item_3 = (grp_item_2 * grp_item_1)

# New DF with formatting
pop_item = pd.DataFrame({
    "Purchase Count": grp_item_1,
    "Item Price": grp_item_2,
    "Total Purchase Value": grp_item_3
})

#Pop & format
pop_item = pop_item.sort_values("Purchase Count", ascending=False) 
pop_item["Item Price"] = pop_item["Item Price"].map("${:.2f}".format)
pop_item["Total Purchase Value"] = pop_item["Total Purchase Value"].map("${:.2f}".format)
pop_item.head()
```
